## EDA

```python
data_2016 = df_2016['Region'].value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x = data_2016.index, y = data_2016.values)
plt.xlabel('Region')
plt.xticks(rotation = 90)
plt.ylabel('Number of Countries')
plt.title('Jumlah negara berdasarkan wilayah', color = 'black', fontsize = 15)
plt.show()
```

dari data diatas negara yang paling banyak berada di wilayah Sub-Saharan Africa

### Correlation Matrix

```python
correlation_matrix = df_2016.select_dtypes(include=['float64', 'int64']).corr()

# Plot heatmap korelasi matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="RdYlGn")
plt.title("Correlation Matrix")
plt.show()
```


Berdasarkan nilai correlation matrix diatas, Economy (GDP percapita) memiliki hubungan positif yang paling kuat terhadap Happiness Score sebesar 0,79 dan diikuti oleh Health dan Family.

### GDP by region

```python
gdp_region = df_2016.groupby('Region')['Economy (GDP per Capita)'].sum()
gdp_region
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP by Region')
plt.ylabel('')
```

Pada diagram lingkaran di atas menyatakan bahwa Central and Eastern Europe    memberikan kontribusi PDB tertinggi, sedangkan North America memberikan kontribusi yang lebih kecil terhadap PDB di tahun 2016.
